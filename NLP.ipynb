Common Vectorization Techniques in NLP:
Bag-of-Words (BoW) Model:

Represents text as a collection (bag) of words, disregarding grammar and word order.
Each document is represented by a vector, where each element is the count or frequency of a word in that document.
Example: For the corpus ["Hello world", "Hello NLP"], the vocabulary is ["Hello", "world", "NLP"], and the BoW vectors are:
Document 1: [1, 1, 0] (1 occurrence of "Hello", 1 of "world", 0 of "NLP")
Document 2: [1, 0, 1] (1 occurrence of "Hello", 0 of "world", 1 of "NLP")
Term Frequency-Inverse Document Frequency (TF-IDF):

Improves upon the BoW model by considering not just the frequency of a word in a document, but also how common or rare the word is across all documents.
TF-IDF score = TF (term frequency)  IDF (inverse document frequency).
This helps emphasize unique words to a document while down-weighting common words (like "the" or "is") across documents.
Word Embeddings (e.g., Word2Vec, GloVe):

Represent words as continuous vectors in a high-dimensional space, capturing semantic relationships.
Unlike BoW and TF-IDF, embeddings retain semantic meaning. For instance, words like "king" and "queen" or "Paris" and "France" have vectors that reflect their relationships (e.g., king - man + woman  queen).
Word2Vec: Uses neural networks to learn word associations and create vectors.
GloVe (Global Vectors): Trained on word co-occurrence statistics across a large corpus.
Doc2Vec:

Extends Word2Vec to represent entire documents as vectors, not just individual words.
Useful for document-level tasks like sentiment analysis and topic modeling.
Count Vectorization:

Similar to BoW, but instead of frequencies, it indicates binary presence (1 if the word is present, 0 if not).

```
def text_preprocess(mess):
    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text
    """
    #check characters to see if they are in punctuation
    nopunc = [char for char in mess if char not in string.punctuation]
    # Join the characters again to form the string.
    nopunc = ''.join(nopunc)
    nopunc = nopunc.lower()

    # Now just remove any stopwords and non alphabets
    nostop=[word for word in nopunc.split() if word.lower() not in
            stopwords.words('english')
     and word.isalpha()]

    return nostop
```

### TF-IDF

TF-IDF stands for **term frequency-inverse document frequency**, and the tf-idf weight is a weight often used in information retrieval and text mining. This weight is a statistical measure used to evaluate how important a word is to a document in a collection or corpus. The importance increases proportionally to the number of times a word appears in the document but is offset by the frequency of the word in the corpus. Variations of the tf-idf weighting scheme are often used by search engines as a central tool in scoring and ranking a document's relevance given a user query.


**TF: Term Frequency**, which measures how frequently a term occurs in a document. Since every document is different in length, it is possible that a term would appear much more times in long documents than shorter ones. Thus, the term frequency is often divided by the document length (aka. the total number of terms in the document) as a way of normalization:

$$ TF(t) = \frac{\text{Number of times term t appears in a document}}{\text{Total number of terms in the document}}. $$
   

**IDF: Inverse Document Frequency**, which measures how important a term is. While computing TF, all terms are considered equally important. However it is known that certain terms, such as "is", "of", and "that", may appear a lot of times but have little importance. Thus we need to weigh down the frequent terms while scale up the rare ones, by computing the following:

$$ IDF(t) = \log_e\Bigg( \frac{\text{Total number of documents}}{\text{Number of documents with term t in it}} \Bigg). $$
